```python
Stats = dataSet.describe()
Stats
```

> ## Checking For missing values in data set

```python
dataSet.isnull().sum()
```

```python
mean_fa = Stats.fixed_acidity.loc['mean']
mean_va = Stats.volatile_acidity.loc['mean']
mean_cl = Stats.chlorides.loc['mean']
mean_fs = Stats.free_sulfur_dioxide.loc['mean']
mean_de = Stats.density.loc['mean']
mean_qu = Stats.quality.loc['mean']
```

```python
newData = dataSet.copy()
```

```python
newData.fixed_acidity = newData.fixed_acidity.fillna(mean_fa)
newData.volatile_acidity = newData.volatile_acidity.fillna(mean_va)
newData.chlorides = newData.chlorides.fillna(mean_cl)
newData.free_sulfur_dioxide = newData.free_sulfur_dioxide.fillna(mean_fs)
newData.density = newData.density.fillna(mean_de)
newData.quality = newData.quality.fillna(mean_qu)
```

```python
newData.isnull().sum()
```

```python
newData
```

```python
newData.info()
```

## Scatter graph

NB: This is based on Fixed acidity against pH

```python
%matplotlib inline
```

```python
mat.figure(figsize = (9,7))
mat.title('Figure acidity over pH',fontdict = {'fontweight' : 'bold'})
mat.scatter(x = 'pH', y = 'fixed_acidity', data = newData, c = 'b', alpha = 0.8)
mat.xlabel('pH', fontsize = 12, fontstyle = 'italic')
mat.ylabel('Fixed acidity')
mat.legend()
```

from the scatterd plot depicting the relationship between the fixed acidity and the PH of red wines, it is evident that those with lower PH levels exibit 
#higher fixed acidity . thus creating a discernible negative slope. in simple terms, as the PH of red wine increases , its fixed acidity tends to decrease.
this implies an inverse correlaton between pH and fixed acidity in the dataset. below is the table being used for the graph :

## histogram 

this is based on the density 

```python
bins = np.arange(0.990, 1.004,0.002)

mat.figure(figsize = (8,7))
mat.hist(newData.density, bins = bins, rwidth = 0.8)
mat.xticks(bins)

mat.title('DENSITY OF THE RED WINES', fontdict = {'fontweight' : 'bold'})
mat.xlabel('Density', fontsize = 12, fontstyle = 'italic')
mat.ylabel('Red Wine', fontsize = 12, fontstyle = 'italic')

mat.show()
```

From the histogram, it can be seen that most red wines had their densities between 0.994 and 1.000. On the other hand, few red wines had their densities between 0.990 and 0.994 and between 1.000 and 1.004. Also, it was gotten from the new Dataset table.

> ## PIE CHART

  NB: This is based on the average amount of residual sugar, sulphates, alcohol, free sulphur dioxide, citric acid and chlorides from the dataset.

```python
newData_stats[['residual_sugar', 'sulphates', 'alcohol', 'free_sulfur_dioxide', 'citric_acid', 'chlorides']].loc['mean']
```

```python
ResidualSugar     = newData_stats['residual_sugar'].loc['mean']
Sulphates         = newData_stats['sulphates'].loc['mean']
Alcohol           = newData_stats['alcohol'].loc['mean']
FreeSulfurDioxide = newData_stats['free_sulfur_dioxide'].loc['mean']
CitricAcid        = newData_stats['citric_acid'].loc['mean']
Chlorides         = newData_stats['chlorides'].loc['mean']

labels = ['Residual sugar', 'Sulphates', 'Alcohol', 'Free sulfur dioxide', 'Citric acid', 'Chlorides']
Values = [0,0,0,0,0,0.6]

mat.pie([ResidualSugar, Sulphates, Alcohol, FreeSulfurDioxide, CitricAcid, Chlorides],
labels = labels, autopct = '%.2f %%', explode = Values)
mat.title('THE COMPONENT OF THE RED WINE', fontdict = {'fontweight' : 'bold'}, ha = 'center')
mat.show()
```

While the detail about the individual qualities of various red wines were given, the analysis employed the mean quality of each component. This approach was taken to streamline the comparison, treating the components as if they belong to a single red wine. The goal was to assess the percentage composition in red wine based on this aggregated perspective.
